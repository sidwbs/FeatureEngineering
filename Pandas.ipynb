## Series
- A Series is a one-dimensional array-like object containing a sequence of values  and an associated array of data labels called its index
- Series is the main data type for Pandas.
- Series is very similar to a Numpy array.
- A series can have axis labels, meaning it can be indexed by a label, instead of just a number location.

## Missing Data
- One of the goals of pandas is to make working with pandas as painless as possible.
- For numeric data pandas uses the floating-point value NAN to represent missing data. This is called a sentinel value that can be detected.
- Up to now, pandas used several values to represent missing data: np.nan
is used for float data, np.nan or None is used for object-dtype data and pd.NaT
for datetime like data.
- A new pd.NA value is introduced to represent scalar missing values. 
- The goal of pd.NA is to provide a 'missing' indicator that can be used consistently across data types

### Cleaning or Editing Strings

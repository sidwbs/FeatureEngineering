### Histogram
- a histogram is a univariate plot which helps us to understand the distribution of a continuous numerical variable. It is used when we have quantitative data.
- it breaks the range of the continuous variabled into intervals of equal length and then counts the number of observations in each interval.
- the way data is distributed or spread out in a histogram is called the distribution.
- One reason why we might use a histogram instead of a bar chart or a dot plot is that we have too many data points.
- the tallest clusters of bars i.e. peaks, in a histogram represent the modes of the data.

To find the optimal number of bins we can use the following rule of thumb:
 $$ binwidth =\frac{(2 * IQR)}{\sqrt[3]{n}} $$ where n = number of rows the dataset

- Then, we obtain bins using the calculated bin-width. $$ bins =\frac{Range}{binwidth} $$

- Kde visualises the distribution of data over a continuous interval.

### Kde Plot
- the kde plot maps an estimate of a probability density function of a random variable. Kde is a fundemental data smoothing problem where inferences about the population are made based on a finite data sample

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns = ['age'])
```

```python
sample_ages.head()
```

```python
sns.kdeplot(data = sample_ages, x = 'age')
plt.show()
```

- Using parameters like clip, bw_adjust

```python
sns.kdeplot(data = sample_ages, x= 'age', clip = [0,100])
plt.show()
```

```python
sns.kdeplot(data = sample_ages, x = 'age', bw_adjust = 0.5)
plt.show()
```

```python
x = np.random.normal(0,1 ,size = 20)
```

```python
from scipy import stats
```

```python
plt.figure(figsize = (8,4), dpi = 200)
bandwidth = x.std() * x.size ** (-0.001)
support = np.linspace(-5,5 ,100)
kernels = []

plt.figure(figsize = (12,6))

for x_i in x:
    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color = 'lightblue')
sns.rugplot(x, color = 'darkblue', linewidth = 4)
plt.show()
```

## Box Plots
- it displays the distribution of continuous variables. It does these through quartiles. We can create a boxplot per category.
- a boxplot, or a box and whisker plot, shows the distribution of numerical data and skewness through displaying the data quantiles
- it is also called as a five-number summary plot, where the five-number summary includes the minimum value, first quartile, median, third quartile and the maximum value
- IQR defines the width of the box. Median is the 50TH Percentile.
- The whiskers are defined by 1.5*IQR
- Outside these whiskers are the Outliers.

```python
sns.boxplot(data = df1, x = 'curb_weight')
plt.show()
```

```python
plt.title('Boxplot: Horsepower')
plt.xlim(30,300)
plt.xlabel('Horsepower')
sns.axes_style('whitegrid')
sns.boxplot(data = df1, x = 'horsepower', color = 'green')
plt.show()
```

- in the boxplot when the median is closer to the left of the box and the whisker is shorter on the left end of the box, we say that the distribution is positively skewed (right skewed)
- Similarly, when the median is closer to the right of the box and the whisker is shorter on the right end of the box, we say that the distribution is negatively skewed (left skewed)

```python
sns.boxplot(data = df1, x = 'price')
plt.show()
```

- from the above plot we can say that the distribution of price is positively skewed

## Violin Plot
- A violin plot allows us to visualise the distribution of a numeric variable for several groups just like box plot
- it is similar to boxplot with a rotated plot on each side, giving more information about the density estimate on the y-axis
- it displays the probability density across the data using KDE

```python
sns.violinplot(data = df1, x = 'horsepower')
plt.show()
```

- The white dot in the plot represents the median
- the thick gray bar in the center represents the interquartile range
- the wider section of the plot represents a higher probability and the skinner sections represent a lower probability for the given value

```python
sns.violinplot(data = df1, x = 'fuel_type', y = 'horsepower', orient = 'v')
plt.show()
```

## Swarm Plot
- a swarmplot simply shows all the data points in the distribution
- for a very large data sets, it wont show all the points but will display the general distribution of them
- it avoids overlapping of points

```python
sns.swarmplot(data = df1, x = 'number_of_doors', y = 'price', )
plt.show()
```

- in a swarmplot dots are arranged in vertical columns
- we can estimate the distribution of our data by looking at the width of the plot

```python
sns.swarmplot(data = df1, x = 'fuel_type', y = 'price', hue = 'number_of_doors')
plt.show()
```

- Using the Parameters dodge, size

```python
sns.swarmplot(data=df1, x = 'fuel_type', y = 'price', hue = 'number_of_doors', dodge = True)
plt.show()
```

```python
sns.swarmplot(data = df2, x = 'math score')
plt.show()
```

```python
sns.swarmplot(data = df2, x = 'math score', size = 2)
plt.show()
```

```python
sns.swarmplot(data = df2, x = 'math score', y = 'race/ethnicity', size = 2)
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.swarmplot(data = df2, x = 'race/ethnicity', y = 'math score', hue = 'gender', dodge = True,size =5)
plt.show()
```

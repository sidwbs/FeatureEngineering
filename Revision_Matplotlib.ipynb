## Matplotlib Figure Object 
- also called OO method
    - using .add_subplot()(Method 1)
    - using .add_axes() (Method 2)
    - Matplotlib sub plots() (Method 3)

```python
#  data
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
print(a)
print(b)
```

```python
x = np.arange(0, 10)
y = 2 ** x
```

```python
print(x)
print(y)
```

```python
# Method 1 -> using .add_subplot()
fig = plt.figure() # creating an empty canvas
ax = fig.add_subplot() # adding axis to the figure we created
ax.plot(x, y) # plotting on the axis using .plot() method
plt.show()
```

```python
fig = plt.figure()  # creating an empty canvas
ax = fig.add_subplot() # adding axes to the figure we created
ax.plot(a, b) # plotting on the axes
plt.show()
```

```python
# Method 2 -> add_axes()

fig = plt.figure() # creating an empty canvas

ax = fig.add_axes([0, 0, 1, 1]) # adding a set of axis to the figure

ax.plot(x, y)
plt.show()
```

```python
fig = plt.figure() # creating an empty canvas

ax = fig.add_axes([0, 0, 1, 1]) # adding a set of axis to the empty canvas
ax.plot(a, b) # plotting on the set of axis
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [1/x1[0], 1/x[1], 1/x[2], 1/x[3]]
```

```python
fig = plt.figure() # creating an empty canvas
ax = fig.add_axes([0, 0, 1, 1]) # adding a set of axis to the figure
ax.plot(x1, y1) # plotting on the set of axis
plt.show()
```

```python
print(a)
print(b)
```

```python
# creating an empty canvas
fig = plt.figure()

# adding two set of axes to the empty canvas
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.2, 0.5, 0.5])

# plotting on the first axes

ax1.plot(a, b)
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_title('Big Figure')

# plotting on the second axes
ax2.plot(a, b)
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_title('Small Figure')

plt.show()
```

```python
# creating a blank canvas
fig = plt.figure()

# adding two axes to the empty canvas
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.5, 0.5])

# plotting on the two axes

ax1.plot(a, b)
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_title('title here')

ax2.plot(a, b)
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_title('title here')
plt.show()
```

```python
# Figure Parameters
fig = plt.figure(figsize = (2, 2), dpi = 200)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(a, b)
plt.show()
```

```python
# Method 3 -> using subplots()

# similar to plt.figure() but this returns a tuple containing figure and axes in the same step
# 2 in 1 step
fig, axes = plt.subplots()

axes.plot(x, y) # plotting on the axes 

axes.set_xlabel('X')

axes.set_ylabel('Y')

axes.set_title('Title here')

plt.show()
```

```python
# option 2

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.hist(np.random.randn(1000))
plt.show()
```

```python
# option 3

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))
axes[1][0].plot(x, x/2)
axes[1][1].scatter(np.random.random(10), np.random.random(10))
axes[0][0].hist(np.random.randn(1000))
```

```python
# setting parameters - color

fig, ax = plt.subplots(figsize = (12, 6), dpi = 200)
ax.plot(x, x+1, color = 'blue', label = 'x+1')
ax.plot(x, x+2, color = 'turquoise', label = 'x+2')
ax.plot(x, x+3, color = 'indianred', label = 'x+3')
ax.legend()
plt.show()
```

```python
# setting parameteres - color, alpha

fig, ax = plt.subplots(figsize = (12, 6), dpi = 200)
ax.set_xlim(0, 20)
ax.set_ylim(0, 100)
ax.plot(x, x**2, color = 'red', alpha = 0.5, label = 'x**2')
ax.plot(x, np.log(x), color = 'turquoise', alpha = 0.5, label = 'log(x)')
ax.plot(x, x+1, color = 'indianred', alpha = 0.5, label = 'x+1')

ax.legend()
plt.show()
```

```python
# setting parameteres - color, linewidth

fig, ax = plt.subplots(figsize = (12, 6), dpi = 200)

ax.plot(x, x+1, color = 'turquoise', linewidth = 0.25)
ax.plot(x, x-2, color = 'navy', linewidth = 0.50)
ax.plot(x, x-3, color = 'indianred', linewidth = 1)
ax.plot(x, x-4, color = 'yellow', linewidth = 8)

plt.show()
```

```python
# setting parameteres - color, linestyles, linewidth

fig, ax = plt.subplots(figsize = (12, 6), dpi = 200)

ax.plot(x, x+1, color = 'turquoise', linewidth = 0.25, linestyle = '--')
ax.plot(x, x-2, color = 'navy', linewidth = 0.50, linestyle = '-')
ax.plot(x, x-3, color = 'indianred', linewidth = 1, linestyle = ':')
ax.plot(x, x-4, color = 'yellow', linewidth = 8, linestyle = '-.')

plt.show()
```

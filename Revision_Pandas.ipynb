# Series
- A series is a data structure in pandas that holds an array of information along with named index.
- Series is built on top of Numpy. 
- A series is a one-dimensional array like object containing a sequence of values and array of data labels called index.
- difference between NUMPY and SERIES is that it can have axis labels, meaning it can be indexed by a label, instead of just a number location.
- It can hold any arbitrary python object.

```python
import numpy as np
import pandas as pd
```

# Useful Methods

```python
# .apply()-> allows us to apply and broadcast custom functions on a dataFrame
def last_four(num):
    return str(num)[-4:]
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
```

```python
df.head()
```

```python
def yelp(price): 
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['Expensive'] = df['total_bill'].apply(yelp)
```

```python
df.head()
```

```python
# apply() with lambda

def simple(num):
    return num * 2
```

```python
lambda num: num * 2
```

```python
df['total_bill'].apply(lambda bill : bill * 0.18)
```

```python
# apply() that uses multiple columns
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
df['Tip Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']),axis=1)
```

```python
# Method 2 -> add_axes()

# creating an empty canvas
fig = plt.figure()

# adding a set of axes to the figure
axes = fig.add_axes([0,0,1,1])

# plotting on the set of axes
axes.plot(x,y)
plt.show()
```

### Adding another set of axes to the figure

```python
# creating an empty canvas
fig = plt.figure()

# adding a set of axes to the figure
axes1 = fig.add_axes([0,0,1,1]) # Large Figure
axes2 = fig.add_axes([0.2,0.2,0.5,0.5]) # Zoomed Figure

# plotting on the Larger of the axes
  
axes1.plot(a,b)
axes1.set_xlabel('x label')
axes1.set_ylabel('y label')
axes1.set_title('Big Figure')

# Plotting on the Zoomed Axes
axes2.plot(a,b)
axes2.set_title('Small Figure')
```

### Moving the 'Small Figure' and edit its parameteres

```python
# Create a blank Canvas
fig = plt.figure()

# adding a set of axis to the figure
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.5,0.5])

# plotting on the Larger axes
axes1.plot(a,b)
axes1.set_xlabel('x label')
axes1.set_ylabel('y label')
axes1.set_title('Small Figure')

# plotting on the Zoomed axes
axes2.plot(a,b)
axes2.set_title('Small figure')
```

```python
# Create a blank Canvas
fig = plt.figure()

# plotting set of axes to the figure
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes3 = fig.add_axes([1,1,0.5,0.5])

# plotting the axes
axes1.plot(a,b)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Big Figure')

axes2.plot(a,b)
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('Zoomed In Figure')

axes3.plot(a,b)

plt.show()
```

### plt.subplots()

```python
# similar to plt.figure() except use tuple unpacking to grab figure and axes
# creating a blank canvas and axes together 
fig, axes = plt.subplots()

# plotting the axes
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
plt.show()
```

### Parameters on subplots()

```python
# creating a blank canvas and axes
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12,8), dpi = 200)

# plotting axis to the fig
axes[0][0].plot(a,b)
axes[0][0].set_title('0 0 Title')

axes[1][1].plot(x,y)
axes[1][1].set_title('1 1 Title')
axes[1][1].set_xlabel('X')

axes[0][1].plot(x,y)
axes[0][1].set_xlabel('X')

axes[1][0].plot(b,a)

# setting parameters at figure level
fig.suptitle('Figure Level', fontsize = 18)
plt.tight_layout()

# using wspace, hspace to add spacing between subplots
fig.subplots_adjust(right = 0.9, left = 0.125, wspace = 0.9, hspace = 0.3)


plt.show()
```

### Setting color parameter

```python
fig, ax = plt.subplots(figsize = (12,6))
ax.plot(x,x+1,color = 'blue', alpha = 0.5) # alpha-> half transparent
ax.plot(x,x+2,color = '#eb9834')
ax.plot(x,x+3,color = '#103ca3')
```

### Linewidth

```python
fig, ax = plt.subplots(figsize = (12,6))

ax.plot(x,x+1,color = 'red', linewidth =0.25)
ax.plot(x,x-2,color = 'red', lw = 0.50)
ax.plot(x,x-3,color = 'red', lw = 1)
ax.plot(x,x-4,color = 'red', lw = 8)
plt.show()
```

### Linestyles

```python
fig, ax = plt.subplots(figsize = (12,6))
ax.plot(x,x-1,color = 'blue', lw = 3, linestyle = '-')
ax.plot(x,x-2,color = 'green', lw = 3, ls = '-.')
ax.plot(x,x-3,color = 'red', lw = 3,ls = '--')
plt.show()
```

### Markers

```python
fig, ax = plt.subplots(figsize = (12,6))
ax.plot(x,x-1,color='red', marker = '+', markersize = 10)
ax.plot(x,x-2,color = 'blue', marker = 'o', markersize =20)
ax.plot(x,x-3,color = 'purple',marker = 's', ms = 20)
ax.plot(x,x-4,color = 'green', marker = 'o', ms = 20)
plt.show()
```

## Bar

```python
nut_butter_prices = {'Almond Butter': 10,
                    'Peanut Butter': 8,
                    'Cashew Butter':12}
fig, ax = plt.subplots()
ax.bar(x = nut_butter_prices.keys(), height=nut_butter_prices.values(), data = nut_butter_prices)
ax.set(title = 'Bar Chart', xlabel = 'Nut_Butter_variety', ylabel = 'Price')
plt.show()
```
